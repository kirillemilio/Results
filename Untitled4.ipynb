```python
N50_dict = {"mutation_rate":[], "coverage":[], "N50":[]}
for i in mutation_per_100:
    for j in coverage_values:
        temp_sum = sum(assembly_data[(assembly_data["mutation_rate"] == i) & 
                                            (assembly_data["coverage"] == j)]["length of path"])
        
        temp_data = list(assembly_data[(assembly_data["mutation_rate"] == i) &
                                            (assembly_data["coverage"] == j)].sort(["length of path"], 
                                                                                  ascending=[0])["length of path"])
        temp_N50_sum = 0
        counter = 0
        N50 = 0
        while(temp_sum / 2 > temp_N50_sum):
            temp_N50_sum += temp_data[counter]
            counter += 1
        N50 = temp_data[counter]
        N50_dict["N50"].append(N50)
        N50_dict["coverage"].append(j)
        N50_dict["mutation_rate"].append(i)
N50_pd = pd.DataFrame(N50_dict)
# N50_pd[N50_pd["mutation_rate"] == 2]
```

```python
g20 = sns.FacetGrid(N50_pd[N50_pd["coverage"] <= 15], col="coverage", margin_titles=True)
g20.map(sns.regplot, "mutation_rate", "N50", scatter_kws={"s": 65})
g20.set_xticklabels(np.array(["", 2, "" , 3, "", 4, "" , 5]))
axes20 = g20.axes
axes20[0,0].set_ylim(0,)
axes20[0,1].set_ylim(0,)
g20.savefig('/home/kirill/Desktop/kirill/TexProjects/Diploma_alt/Plots_for_presentation/N50_over_mutation_rate_part_1.png',bbox_inches="tight")

g30 = sns.FacetGrid(N50_pd[N50_pd["coverage"] > 15], col="coverage", margin_titles=True)
g30.map(sns.regplot, "mutation_rate", "N50", scatter_kws={"s": 65})
g30.set_xticklabels(np.array(["", 2, "" , 3, "", 4, "" , 5]))
axes30 = g30.axes
axes30[0,0].set_ylim(0,)
axes30[0,1].set_ylim(0,)
g30.savefig('/home/kirill/Desktop/kirill/TexProjects/Diploma_alt/Plots_for_presentation/N50_over_mutation_rate_part_2.png',bbox_inches="tight")
```

```python
error_samples = assembly_data[assembly_data["mismatch"] != 0]
correct_samples = assembly_data[assembly_data["mismatch"] == 0]

positive_dict = {"mutation_rate":[], "coverage":[], "number of correct paths":[]}
for i in mutation_per_100:
    for j in coverage_values:
        positive_dict["number of correct paths"].append(correct_samples[(correct_samples["mutation_rate"] == i) & 
                                            (correct_samples["coverage"] == j)].shape[0])
        positive_dict["coverage"].append(j)
        positive_dict["mutation_rate"].append(i)

negative_dict = {"mutation_rate":[], "coverage":[], "number of correct paths":[]}
for i in mutation_per_100:
    for j in coverage_values:
        negative_dict["number of correct paths"].append(error_samples[(error_samples["mutation_rate"] == i) & 
                                            (error_samples["coverage"] == j)].shape[0])
        negative_dict["coverage"].append(j)
        negative_dict["mutation_rate"].append(i)

positive_proportion_dict = {"mutation_rate":[], "coverage":[], "fraction of correct paths":[]}
for i in mutation_per_100:
    for j in coverage_values:
        positive_proportion_dict["fraction of correct paths"].append(float(correct_samples[(correct_samples["mutation_rate"] == i) & 
                                            (correct_samples["coverage"] == j)].shape[0]) / assembly_data[(assembly_data["mutation_rate"] == i) & 
                                            (assembly_data["coverage"] == j)].shape[0])
        positive_proportion_dict["coverage"].append(j)
        positive_proportion_dict["mutation_rate"].append(i)

positive_proportion_pd = pd.DataFrame(positive_proportion_dict)
```

```python
g9 = sns.FacetGrid(positive_proportion_pd, col="mutation_rate", margin_titles=True)
g9.map(sns.regplot, "coverage", "fraction of correct paths", fit_reg=False, scatter_kws={"s": 65})
g9.savefig('/home/kirill/Desktop/kirill/TexProjects/Diploma_alt/Plots_for_presentation/correct_paths_fraction_coverage.png',bbox_inches="tight")
```

```python
positive_proportion_vertexies_dict = {"mutation_rate":[], "coverage":[], "fraction of correct vertexies":[]}
for i in mutation_per_100:
    for j in coverage_values:
        positive_proportion_vertexies_dict["fraction of correct vertexies"].append(1.0 - sum(assembly_data[(assembly_data["mutation_rate"] == i) & 
                                            (assembly_data["coverage"] == j)]["mismatch"]) / 
                                        float(sum(assembly_data[(assembly_data["mutation_rate"] == i) & 
                                            (assembly_data["coverage"] == j)]["number of vertexies"])))
        positive_proportion_vertexies_dict["coverage"].append(j)
        positive_proportion_vertexies_dict["mutation_rate"].append(i)

positive_proportion_vertexies_pd = pd.DataFrame(positive_proportion_vertexies_dict)
```
