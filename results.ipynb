```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_new_table(table):
    read_length_list = table_alt.read_length.unique()
    error_rate_list = table_alt.error_rate.unique()
    mutation_rate_list = table_alt.mutation_rate.unique()
    l = []
    my_df = pd.DataFrame(columns = table.columns)
    for r in read_length_list:
        for e in error_rate_list:
            for m in mutation_rate_list:
                result = table[(table.error_rate == e) & (table.read_length == r) & (table.mutation_rate == m)]
                result = result[result.N50 == result.N50.max()]
                if(not result.empty):
                    result = result[result.kmer == result.kmer.max()]
#                     print(result)
                    my_df = my_df.append(result)
    return my_df

table = pd.read_csv('final_table.csv')
table_alt = table.iloc[:, [1, 2, 3, 4, 13]]
my_df = get_new_table(table_alt)
```

Графики зависимости N50 от длины рида и количества ошибок на 100 нуклеотидов при количестве мутаций на 100 нуклеотидов равным 5(ближайшее к среднему).
Сборки, для которых количество ошибок на 100 нуклеотидов(error_rate) равно 0 пришлосm вынести на отдельный график, так как для них N50 намного больше, чем у остальных сборок.

```python
my_df2 = my_df[(my_df.mutation_rate == 5) & (my_df.error_rate != 0)].sort(['read_length', 'error_rate'], 
                                           ascending=[1, 1]).loc[:, ['read_length', 'error_rate', 'N50']]
g = sns.FacetGrid(my_df2, col='error_rate')
g = g.map(plt.scatter, "read_length", "N50", edgecolor="w")
```
